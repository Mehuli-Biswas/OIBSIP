```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set()
```

#****Unemployment in India(2019-2020)****

```python
df=pd.read_csv("Unemployment in India.csv")
df.head(30)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.columns
```

```python
#Removing whitespaces from column names
df.columns=df.columns.str.strip()
df.columns
```

```python
#Checking for the missing values
df.isnull().sum()
```

```python
#Dropping records with missing values
df.dropna(inplace=True)
df.isnull().sum()
```

```python
#Checking for duplicated records
df.duplicated().sum()
```

```python
sns.heatmap(df.corr(),annot=True)
plt.title('Correlation Heatmap for unemployment in India (2019-2020)',fontsize=20)
```

```python
sns.pairplot(df)
```

```python
df['Area'].value_counts()
```

```python
plt.pie(df['Area'].value_counts(),labels=df['Area'].value_counts().index,autopct='%1.1f%%')
```

```python
avg_uemp_rate=df.groupby(df['Region'])['Estimated Unemployment Rate (%)'].mean()
plt.figure(figsize=(12,6))
sns.barplot(x=avg_uemp_rate.index, y=avg_uemp_rate.values)
plt.title('Region vs Average Unemployment Rate\n2019-2020',fontsize=20)
plt.xlabel('Region')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.xticks(rotation=90)

print(f'{avg_uemp_rate.idxmax()} had the highest average estimated Unemployment rate of {avg_uemp_rate.max():0.2f}%')
print(f'{avg_uemp_rate.idxmin()} had the lowest average estimated Unemployment rate of {avg_uemp_rate.min():0.2f}%')
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(x='Date',y='Estimated Unemployment Rate (%)',hue='Region',data=df)
plt.title('Variation of estimated unemployment rate in different regions over time\n2019-2020',fontsize=20)
plt.xlabel('Data')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.xticks(rotation=45)
plt.legend(loc='center left', bbox_to_anchor=(1.1, 0.5))
```

```python
plt.figure(figsize=(12,6))
sns.set()
sns.scatterplot(x='Estimated Employed',y='Estimated Unemployment Rate (%)',hue='Region',data=df)
plt.title('Estimated Employed vs Estimated Unemployment Rate (%)\n(2019-2020)',fontsize=20)
plt.xlabel('Estimated Employed')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.legend(loc='center left',bbox_to_anchor=(1.1,0.5))
```

```python
plt.figure(figsize=(12,6))
sns.set()
sns.scatterplot(x='Estimated Labour Participation Rate (%)',y='Estimated Unemployment Rate (%)',hue='Region',data=df)
plt.title('Estimated Labour Participation Rate (%) vs Estimated Unemployment Rate (%)\n(2019-2020)',fontsize=20)
plt.xlabel('Estimated Employed')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.legend(loc='center left',bbox_to_anchor=(1.1,0.5))
```

```python
sns.heatmap(df1.corr(),annot=True)
plt.title('Correlation Heatmap for unemployment of India\n(2020-)',fontsize=20)
```

```python
avg_unemp_rate=df1.groupby(df1['Region'])['Estimated Unemployment Rate (%)'].mean()
plt.figure(figsize=(12,6))
sns.barplot(x=avg_unemp_rate.index, y=avg_unemp_rate.values,palette='Dark2_r')
plt.title('Region vs Average Unemployment Rate\n(In 2020)',fontsize=20)
plt.xlabel('Region')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.xticks(rotation=90)
avg_unemp_rate.max()
avg_unemp_rate.idxmax()

print(f'{avg_unemp_rate.idxmax()} had the highest average estimated Unemployment rate of {avg_unemp_rate.max():0.2f}%')
print(f'{avg_unemp_rate.idxmin()} had the lowest average estimated Unemployment rate of {avg_unemp_rate.min():0.2f}%')
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(x='Date',y='Estimated Unemployment Rate (%)',hue='Region',data=df1)
plt.title('Variation of estimated unemployment rate in different regions over time\n(In 2020)',fontsize=20)
plt.xlabel('Data')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.xticks(rotation=45)
plt.legend(loc='center left', bbox_to_anchor=(1.1, 0.5))
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(x='Estimated Employed',y='Estimated Unemployment Rate (%)',hue='Region',data=df1)
plt.title('Estimated Employed vs Estimated Unemployment Rate (%)',fontsize=20)
plt.xlabel('Estimated Employed')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.legend(loc='center left',bbox_to_anchor=(1.1,0.5))
```

```python
plt.figure(figsize=(12,6))
sns.set()
sns.scatterplot(x='Estimated Employed',y='Estimated Unemployment Rate (%)',hue='Region.1',data=df1)
plt.title('Estimated Employed vs Estimated Unemployment Rate (%)\n(2020-)',fontsize=20)
plt.xlabel('Estimated Employed')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.legend(loc='center left',bbox_to_anchor=(1.1,0.5))
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(x='Estimated Labour Participation Rate (%)',y='Estimated Unemployment Rate (%)',hue='Region.1',data=df1)
plt.title('Estimated Labour Participation Rate (%) vs Estimated Unemployment Rate (%)\n(2020-)',fontsize=20)
plt.xlabel('Estimated Employed')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.legend(loc='center left',bbox_to_anchor=(1.1,0.5))
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(x='Estimated Labour Participation Rate (%)',y='Estimated Unemployment Rate (%)',hue='Region',data=df1)
plt.title('Estimated Labour Participation Rate (%) vs Estimated Unemployment Rate (%)\n(2020-)',fontsize=20)
plt.xlabel('Estimated Employed')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.legend(loc='center left',bbox_to_anchor=(1.1,0.5))
```

#****Conclusion****
- Covid-19 was a difficult time, during which many people were rendered jobless.
- In India, an unprecedented surge in unemployment rate was recorded in the months of March, April and   May of the year 2020 in major states or regions of the country.
- The highest average unemployment rate recorded was 28.35% in Tripura.
- The second highest average unemployment rate recorded was 27.48% in Haryana in the same year.
